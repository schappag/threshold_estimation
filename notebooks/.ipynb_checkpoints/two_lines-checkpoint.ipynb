# Fit two linear segments to data
### **Code Summary**  
- The dataset is divided into **two continuous segments** by identifying **one breakpoint (inflection point)** where the trend changes.  
- **Breakpoint Selection:** Uses an **error-minimization approach** to determine the optimal single breakpoint.
- **Approach:** Instead of fitting separate regressions to each segment, a **spline-based approach** is used to ensure continuity at the breakpoint.  
- **Repetitions:** The process is repeated **50 times**, with each iteration following these steps:  
  - **A random selection of 80% of the data** is taken to perform the calculations.  
  - The breakpoint is determined independently for each subset.  
  - If a **true breakpoint exists**, its estimated location will remain stable across repetitions.  
  - If no true breakpoint exists, results will vary, indicating the algorithm is identifying **local minima** instead of a real structural change.  
- **Final Output:** The **mean breakpoint position** across all repetitions and its corresponding power values are reported.  


```python
import numpy as np
import pandas as pd
from scipy.interpolate import interp1d
from scipy.signal import decimate
import matplotlib.pyplot as plt

def two_lines(t, y, delta_l, delta_u):
    It = (t >= t[0] + delta_l) & (t < t[-1] - delta_u)
    t_fit, y = t[It], y[It]
    y_fit, threshold_time = perform_fit(t_fit, y)
    return t_fit, y_fit, threshold_time

def perform_fit(t, y):
    error_matrix = calculate_error_matrix(t, y)
    a = np.nanargmin(error_matrix)
    y_fit, c = linreg(t, y, a)
    threshold_time = np.array([t[a]])
    return y_fit, threshold_time

def calculate_error_matrix(t, y):
    len_t = len(t)
    error_matrix = np.full(len_t, np.nan)

    for i in range(len_t - 1):
        y_fit, _ = linreg(t, y, i)
        error_matrix[i] = np.linalg.norm(y - y_fit) ** 2

    return error_matrix

def linreg(t, y, i):
    if invalid_indices(i, t):
        return np.full_like(t, 1e10), np.array([1, 2])
    k = np.array([t[0], t[i], t[-1]])
    phi = create_spline_matrix(t, k)
    c = np.linalg.lstsq(phi, y, rcond=None)[0]
    y_fit = np.dot(phi, c)
    return y_fit, c

def invalid_indices(i, t):
    if len(t) == 0:
        return True
    t_min = int(len(t) * 0.2)
    t_max = int(len(t) * 0.8)
    return (i == 0 or i == len(t) - 1 or i <= t_min or i >= t_max)

def create_spline_matrix(t, k):
    return np.column_stack([
        1 - np.maximum(0, np.minimum(1, (t - k[0]) / (k[1] - k[0]))),
        np.maximum(0, np.minimum(1, (t - k[0]) / (k[1] - k[0]))) - np.maximum(0, np.minimum(1, (t - k[1]) / (k[2] - k[1]))),
        np.maximum(0, np.minimum(1, (t - k[1]) / (k[2] - k[1])))
    ])

def perform_fit_2lines(time, meas, delta_l, delta_u):
    split_point = round(len(time) * 0.8)
    time_train, meas_train = time[:split_point], meas[:split_point]
    time_test, meas_test = time[split_point+1:-1], meas[split_point+1:-1]

    sort_idx_train = np.argsort(time_train)
    time_train, meas_train = time_train[sort_idx_train], meas_train[sort_idx_train]

    tr, yr, tc = two_lines(time_train, meas_train, delta_l, delta_u)

    meas_pred = interp1d(tr, yr, kind='linear', fill_value='extrapolate')(time_test)
    err = meas_test - meas_pred
    mse = np.mean(err**2) / len(time_test)
    
    return mse, np.sort(tc)

def inflection_point(filename, titlename, runs):
    data = pd.read_csv(filename)
    
    time_seconds = data.iloc[:, 0].values
    signal = data.iloc[:, 1].values

    time_minutes = time_seconds / 60
    time_minutes -= min(time_minutes)
    n_dec = 10
    
    filtered_signal = decimate(signal, n_dec, ftype='iir', zero_phase=False)
    filtered_time = decimate(time_minutes, n_dec, ftype='iir', zero_phase=False)
    delta_l, delta_u = 0.1, 0.1

    MSE = np.zeros(runs)
    time_threshold_pred = np.zeros(runs)

    for rep in range(runs):
        p = np.random.permutation(len(filtered_time))

        if len(p) < 3:
            continue

        mse, time_threshold = perform_fit_2lines(filtered_time[p], filtered_signal[p], delta_l, delta_u)

        if len(time_threshold) != 1:
            time_threshold = np.array([filtered_time[0]])

        time_threshold_pred[rep] = time_threshold
        MSE[rep] = mse

    mean_time_threshold = np.mean(time_threshold_pred)

    fig, axes = plt.subplots(2, 1, figsize=(5,4), gridspec_kw={'height_ratios': [3, 1]})

    axes[0].plot(time_minutes, signal, label='Original Signal', color='gray', alpha=0.5)
    axes[0].plot(filtered_time, filtered_signal, label='Filtered Signal', color='black', linewidth=1)

    for time_threshold in time_threshold_pred:
        axes[0].axvline(time_threshold, color='blue', linestyle='--', alpha=0.3)

    axes[0].axvline(mean_time_threshold, color='red', linestyle='-', linewidth=2, label='Threshold (mean)')

    axes[0].grid(visible=True, which='both', linestyle='--', linewidth=0.5)
    axes[0].minorticks_on()
    axes[0].set_xlabel('Time (minutes)')
    axes[0].set_ylabel('Filtered Signal')
    axes[0].set_title('Signal with Estimated Threshold')

    axes[1].hist(time_threshold_pred, bins=20, color='blue', alpha=0.7, edgecolor='black')
    axes[1].axvline(mean_time_threshold, color='red', linestyle='-', linewidth=2)
    axes[1].set_xlabel('Estimated Threshold (minutes)')
    axes[1].set_ylabel('Frequency')
    axes[1].set_xlim([np.floor(np.min(time_threshold_pred)), np.ceil(np.max(time_threshold_pred))])
    fig.legend(loc='upper center', bbox_to_anchor=(0.5, 1.15), ncol=3, frameon=False)

    plt.tight_layout()


    return pd.DataFrame({'Threshold': [mean_time_threshold], 
                         'Median_MSE': [np.median(MSE)]})
```

# No true breakpoint

```python
import pandas as pd

filename = "../data/simulated_data_no_breakpoints.csv" 
titlename = "Test_No_True_Breakpoints_Two_Lines"

results = inflection_point(filename, titlename, 50)

print("Inflection Point Detection Results:")
print(results)
```
